```python
import pandas as pd
import json
import numpy as np
```

```python
with open('data/sankey.json', 'r') as fh:
    data = json.load(fh)
```

```python
flattened = []
for path in data:
    major1 = path['major1']
    major1_cnt = path['major1_cnt']
    
    for major2, major2_cnt in path['major2'].items():
        for minor1, minor1_cnt in path['minor1'].items():
            for minor2, minor2_cnt in path['minor2'].items():
                flattened.append({
                    'major1': major1,
                    'major1_cnt': major1_cnt,
                    'major2': major2,
                    'major2_cnt': major2_cnt,
                    'minor1': minor1,
                    'minor1_cnt': minor1_cnt,
                    'minor2': minor2,
                    'minor2_cnt': minor2_cnt,
                })
```

```python
df = pd.DataFrame(flattened)
df
```

```python
df.to_csv('data/sankey.csv')
```
